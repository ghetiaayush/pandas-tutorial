```python
import pandas as pd 
import numpy as np 
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
df1
```

```python
df.join(df1, lsuffix='_caller', rsuffix='_df1')
```

```python
df.set_index('key').join(df1.set_index('key'))
```

```python
df.join(df1.set_index('key'), on='key')
```
